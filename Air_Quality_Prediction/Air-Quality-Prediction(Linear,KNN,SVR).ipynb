```python
# Separate input features and target variable
X=df.drop(columns=['Date','Time','CO(GT)']) # Independent variables
Y = df['CO(GT)'] # Target variable
```

```python
X.shape # Check the shape of feature matrix X (rows = samples, columns = features)
```

```python
Y.shape # Check the shape of target variable Y
```

```python
from sklearn.model_selection import train_test_split # Import function to split data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3) # Split the dataset into training (70%) and testing (30%) sets
```

```python
from sklearn.linear_model import LinearRegression # Import Linear Regression model from scikit-learn
```

```python
model = LinearRegression() # Initialize Linear Regression model
model.fit(X_train, Y_train) # Train Linear Regression model on training data
```

```python
model.predict([[1300, 85, 7.2, 900, 120, 1100, 115, 1500, 1050, 13.5, 60.2, 0.75]]) # Predict CO(GT) for a new sample (feature names not passed; warning is harmless)
```

```python
model.coef_  # Coefficients of the trained Linear Regression model
# model.coef_ gives the weight for each feature in X
```

```python
# Manually calculate predicted Y (CO concentration) for a new sample
# Formula: Y = (coef1 * feature1) + (coef2 * feature2) + ... + (coef12 * feature12)
Y = (
    (1.43948106e-03 * 1300) 
    + (8.28934540e-04  * 85) 
    + (1.51744570e-01  * 7.2) 
    + (-1.32175558e-03 * 900) 
    + (4.31434361e-03  * 120) 
    + (-1.53658138e-04 * 1100)
    + (4.70209629e-03  * 115) 
    + (-4.61220634e-04 * 1500) 
    + ( -5.78848880e-04 * 1050) 
    + (-4.75589733e-02  * 13.5) 
    + (-1.13174958e-02  * 60.2) 
    + (7.92219299e-01 * 0.75)
)

Y + model.intercept_ # Add the intercept of the model to get final predicted Y
```

```python
from sklearn.metrics import mean_absolute_error # Import MAE metric for evaluation
pred = model.predict(X_test) # Predict CO(GT) values for test set using Linear Regression
mean_absolute_error(Y_test,pred) # Calculate MAE to evaluate Linear Regression performance
```

```python
import matplotlib.pyplot as plt # Import matplotlib for visualization
import numpy as np # Set figure size for plot

# Define function to classify CO levels into air quality categories
def air_quality_category(co):
    if co <= 4.0:
        return "Good"
    elif co <= 10.0:
        return "Moderate"
    else:
        return "Unhealthy"

# Create actual and predicted categories
actual_category = [air_quality_category(co) for co in Y_test.values]
pred_category = [air_quality_category(co) for co in pred]

# Map categories to colors
color_map = {"Good": "green", "Moderate": "orange", "Unhealthy": "red"}
actual_colors = [color_map[cat] for cat in actual_category]
pred_colors = [color_map[cat] for cat in pred_category]

plt.figure(figsize=(14,6))

# Plot category background bands
plt.fill_between(range(len(Y_test)), 0, 4, color='green', alpha=0.1)
plt.fill_between(range(len(Y_test)), 4, 10, color='orange', alpha=0.1)
plt.fill_between(range(len(Y_test)), 10, max(max(Y_test.values), max(pred)), color='red', alpha=0.1)

# Plot actual and predicted points
plt.scatter(range(len(Y_test)), Y_test.values, color=actual_colors, label="Actual CO(GT)", alpha=0.7, s=20)# Plot actual CO(GT) levels
plt.scatter(range(len(pred)), pred, color=pred_colors, label="Predicted CO(GT)", alpha=0.7, s=20, marker='x') # Plot predicted CO(GT) levels with 'x' marker

plt.xlabel("Test Sample Index") # Label x-axis
plt.ylabel("CO(GT) Level") # Label y-axis
plt.title("Linear Regression: Actual vs Predicted CO(GT) with Air Quality Categories")# Set plot title
plt.legend()# Show legend
plt.grid(True, linestyle='--', alpha=0.5)
plt.show() # Display plot
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Function to classify CO levels
def air_quality_category(co):
    if co <= 4.0:
        return "Good"
    elif co <= 10.0:
        return "Moderate"
    else:
        return "Unhealthy"

# Convert actual CO values to categories
Y_test_cat = [air_quality_category(co) for co in Y_test]

# Convert predictions to categories
pred_lr_cat = [air_quality_category(co) for co in model.predict(X_test)]
pred_knn_cat = [air_quality_category(co) for co in knnModel.predict(X_test)]
pred_svr_cat = [air_quality_category(co) for co in svr_model.predict(X_test)]

# All possible classes
classes = ["Good", "Moderate", "Unhealthy"]

# Linear Regression
print(" Linear Regression Classification Report ")
print(classification_report(Y_test_cat, pred_lr_cat, labels=classes, zero_division=0))
print("Accuracy:", accuracy_score(Y_test_cat, pred_lr_cat))
print("Confusion Matrix:\n", confusion_matrix(Y_test_cat, pred_lr_cat, labels=classes))

# KNN
print("\n KNN Regression Classification Report ")
print(classification_report(Y_test_cat, pred_knn_cat, labels=classes, zero_division=0))
print("Accuracy:", accuracy_score(Y_test_cat, pred_knn_cat))
print("Confusion Matrix:\n", confusion_matrix(Y_test_cat, pred_knn_cat, labels=classes))

# SVR
print("\nSVR Regression Classification Report ")
print(classification_report(Y_test_cat, pred_svr_cat, labels=classes, zero_division=0))
print("Accuracy:", accuracy_score(Y_test_cat, pred_svr_cat))
print("Confusion Matrix:\n", confusion_matrix(Y_test_cat, pred_svr_cat, labels=classes))
```
